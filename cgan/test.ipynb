```python
import os
import requests
import pandas as pd
import time

# Create a folder to save the images
save_folder = 'images'
os.makedirs(save_folder, exist_ok=True)

# Check if the CSV file already exists
csv_exists = os.path.exists('df_with_images.csv')

# Define the headers for the CSV
headers = ['ID', 'Style', 'agreeableness', 'anger', 'anticipation', 'arrogance',
           'disagreeableness', 'disgust', 'fear', 'gratitude', 'happiness',
           'humility', 'love', 'optimism', 'pessimism', 'regret', 'sadness',
           'shame', 'shyness', 'surprise', 'trust', 'neutral', 'Image URL', 'Image Path']

# Create a new DataFrame with the appropriate columns and headers if the CSV doesn't exist
if not csv_exists:
    data = {header: [] for header in headers}
    df = pd.DataFrame(data)

    # Save the DataFrame to a CSV file with headers
    df.to_csv('df_with_images.csv', index=False, header=headers)

# Check the CSV for previously downloaded images
if os.path.exists('df_with_images.csv'):
    existing_df = pd.read_csv('df_with_images.csv')
    downloaded_urls = existing_df['Image URL'].tolist()
    new_df_emo = new_df_emo[~new_df_emo['Image URL'].isin(downloaded_urls)]

# Function to download and save an image to the folder
def download_and_save_image(image_url, save_folder):
    image_name = os.path.basename(image_url)
    file_path = os.path.join(save_folder, image_name)

    if os.path.exists(file_path):
        return file_path

    max_retries = 10  # Set the maximum number of retries
    retries = 0
    while retries < max_retries:
        response = requests.get(image_url)
        if response.status_code == 200:
            image_name = os.path.basename(image_url)
            file_path = os.path.join(save_folder, image_name)
            with open(file_path, 'wb') as f:
                f.write(response.content)

            # Append the new row to the CSV
            row = new_df_emo[new_df_emo['Image URL'] == image_url].copy()
            row['Image Path'] = str(file_path)
            row.to_csv('df_with_images.csv', mode='a', header=False, index=False)

            return file_path  # Return the file path after a successful download

        retries += 1
        if retries >= max_retries:
            # Skip the URL after reaching the maximum number of retries
            return None  # Return None to indicate a failed download
        # Wait for a while before the next retry (e.g., 5 seconds)
        time.sleep(5)

# Limit the number of images to download

for url in new_df_emo['Image URL']:

    download_and_save_image(url, save_folder)
```

```python
# Define the function to alter an image based on a mood
def alter_image_with_mood(generator, image_path, mood, latent_dim, selected_columns):
    # Load and preprocess the input image
    input_image = preprocess_image(image_path)
    
    # Generate a random noise vector
    noise = np.random.normal(0, 1, (1, latent_dim))
    
    # Prepare the mood probabilities based on the selected mood
    mood_probabilities = np.zeros(len(selected_columns) - 1)
    mood_probabilities[selected_columns.index(mood)] = 1.0
    
    # Combine noise and mood probabilities
    input_data = np.concatenate([noise, mood_probabilities.reshape(1, -1)], axis=1)
    
    # Generate an altered image
    altered_image = generator.predict(input_data)
    
    return altered_image

# Example usage:
altered_image = alter_image_with_mood(generator, 'input_image.jpg', 'happiness', latent_dim, selected_columns)
```
